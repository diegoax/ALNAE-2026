Vamos a ver uno de los usos más clásicos de la SVD en estadística: el análisis de componentes principales.

Supongamos que tenemos $X \in \mathbb{R}^{n\times d}$ una matriz con datos, donde $n$ es la cantidad de individuos y $d$ la cantidad de variables.

En este ejemplo, supongamos que tenemos $d=2$ variables, la matriz sería algo así:

$$ X = \begin{pmatrix}
X_{11} & X_{12} \\
X_{21} & X_{22} \\
\vdots & \vdots \\
X_{n1} & X_{n2}
\end{pmatrix}
$$

Cada fila de la matrix, es un vector $(X_{i1}, X_{i2})$ que se puede representar en un plano, y es el gráfico que hay más abajo.

```r
n <- 1000
x <- rnorm(n)
e <- rnorm(n, sd = 1)
y <- 2*x + e

X <- cbind(x,y)
X <- scale(X, center = TRUE, scale = FALSE)
plot(X, asp = 1, pch = 20)
```

Si consideramos la SVD de $X$, y definimos las matrices de rango 1, $\tilde X_1 = \sigma_1 u_1 v_1^\top$ y $\tilde X_2 = \sigma_2 u_2 v_2^\top$, tenemos que $X = \tilde X_1 + \tilde X_2$.

Veamos que hay en cada una de estas matrices de rango 1, la asociada a $\sigma_1$:

```r
SVD <- svd(X)
U <- SVD$u
D <- SVD$d
V <- SVD$v

plot(X, asp = 1, pch = 20)
points(tcrossprod(U[,1], V[,1])*D[1], col = 2)
```

La asociada a $\sigma_2$

```r
plot(X, asp = 1, pch = 20)
points(tcrossprod(U[,2], V[,2])*D[2], col = 2)
```
